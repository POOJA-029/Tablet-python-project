```python
import requests
from bs4 import BeautifulSoup
import pandas as pdbbb
```

```python
page = requests.get("https://www.ebay.com/e/row/tabletsrefurbishedrow")
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
page.status_code
```

```python
name_tags=soup.select(".s-item__title")
name=[]
for i in name_tags:
    name.append(i.get_text())
print(name)
```

```python
len(name)
```

```python
price_tags = soup.select(".s-item__price")
price = []
for tag in price_tags:
    text = tag.get_text()
    cleaned_text = text.replace('$', '').strip()  # Remove the Rupee symbol and extra whitespace
    price.append(cleaned_text)
    print("Prices:", price)
```

```python
len(price)
```

```python
shipping_tags = soup.select(".s-item__shipping.s-item__logisticsCost")
shipping = []
for tag in shipping_tags:
    text = tag.get_text()
    cleaned_text = text.replace('$', '').strip()  # Remove the dollar symbol and extra whitespace
    shipping.append(cleaned_text)

print("shipping:", shipping)
```

```python
len(shipping)
```

```python
eBay_Refurbished_tags=soup.select(".s-item__certified-refurbished.s-item__certified-refurbished--isLarge")
eBay_Refurbished=[]
for i in eBay_Refurbished_tags:
    eBay_Refurbished.append(i.get_text())
print(eBay_Refurbished)
```

```python
len(eBay_Refurbished)
```

```python
import pandas as pd
tab = pd.DataFrame({
    "Tab_Name": name,
    "E-side": eBay_Refurbished,
    "shipping_cost": shipping,
    "Tab_price":price
})
tab
```

```python
tab.to_csv("tab_Report_Scraping.csv",index=False)
```
